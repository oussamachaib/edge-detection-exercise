## 1) Image structure
We will start by opening one of the PLIF demo images provided in the "data" folder. Most experimental images are acquired in single-channel (grayscale) with 8- to 16-bit depths being the most common. In this lab, we will consider 8-bit images.

```python
img_id = "sweeney1.png"
img = cv2.imread(f"data/{img_id}",0)
img = img.astype(np.uint8)
```

Every pixel in the image should contain an intensity $i$ within the dynamic range $[0,\,2^8-1]$:
\begin{align}
i \in [0,\,255]
\end{align}
The image structure can be better visualized by plotting the image histogram which represents the unnormalized frequency of each intensity value $i$ within the image.

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 3))

plt.subplot(121)
plt.imshow(img, cmap='RdBu_r', vmin=0, vmax=255)
plt.colorbar(fraction=0.035, pad=0.04,ticks=[0, 255])
plt.xlabel("x [px]")
plt.ylabel("y [px]")
plt.title(img_id)

plt.subplot(122)
plt.hist(img.flatten(), bins = np.arange(1,256,1),color='grey',alpha = 1)
plt.xlabel("Intensity, $i$ [-]")
plt.ylabel("Count [-]")

plt.rcParams.update({'font.size': 12})

plt.show()
```

Often, PLIF image histograms will display a bimodal distribution corresponding the reactant (low intensities) and product (high intensities) regions. The variance of the latter is typically larger due to higher noise intensity on the product side.

## 2) Visualizing the image histogram
A way to get an intuitive approach to the histogram shown previously is to apply a thresholding operation the image. Thresholding, in this context, is akin to a high-pass filter which only accepts intensities larger than the threshold specified.
\
\
By changing the value of the threshold (dashed red line in the histogram), one can "segment" the image into two regions: a foreground (brightest pixels) and a background (darker pixels). This operation is often referred to as binarization in literature: foreground pixels (burnt region) are assigned values of ones and background pixels (unburnt region) are assigned zeros. This intuitive visualization of the histogram will prove useful when we discuss segmentation as a flame front detection approach (notice how setting the threshold to a value close to 100 does a decent job at isolating the unburnt and burnt regions).

```python
wg.interact(fun.segmentation, img0 = wg.fixed(img), thresh = (1,255,round(255/10)))
```

-----
## 2) Locating a flame front
From the previous example, we've noticed that binarization is very sensitive to the choice of threshold. By observation, values close to 100 tend to do a good job at isolating both burnt and unburnt regions. For this example, we will binarize the image by setting the threshold to $\text{thresh} = 105$. To select the flame front from the binarized image, highlighting the perimeter is sufficient. In this example, we highlight the perimeter of the image by making use of erosion using a $3\times3$ box kernel $H$.
\
\
The OpenCV documentation provides more details on erosion (if needed): https://docs.opencv.org/3.4/db/df6/tutorial_erosion_dilatation.html

```python
# Thresholded image
thresh_manual = 105

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 3))
plt.subplot(121)
img1 = img>thresh_manual
img1 = (img1).astype(int)
plt.imshow(img1, cmap='gray', vmin=0, vmax=1)
plt.colorbar(fraction=0.035, pad=0.04,ticks=[0, 1])
plt.xlabel("x [px]")
plt.ylabel("y [px]")
plt.title(img_id)

# Perimeter of thresholded image
img2 = img1.astype(np.uint8)
H = (1/9)*np.ones((3,3))
img_erode = cv2.erode(img2,kernel = H)
contour_manual = img1-img_erode
plt.subplot(122)
plt.imshow(contour_manual, cmap='gray')
plt.colorbar(fraction=0.035, pad=0.04,ticks=[0, 1])
plt.xlabel("x [px]")
plt.ylabel("y [px]")
plt.title("Contour")
plt.rcParams.update({'font.size': 12})

plt.show()
```

The contour can then be overlayed on the original image for better visibility.

```python
# Locating the coordinates of flame front pixels
y_manual,x_manual = np.where(contour_manual == 1)
plt.figure(figsize=(5, 5))
plt.imshow(img, cmap='RdBu_r', vmin=0, vmax=255)
plt.colorbar(fraction=0.035, pad=0.04,ticks=[0, 255])
plt.xlabel("x [px]")
plt.ylabel("y [px]")
plt.title(img_id)
plt.plot(x_manual,y_manual,'.k',markersize = 1, label = "Contour")
plt.legend()
plt.rcParams.update({'font.size': 12})
plt.show()
```

## 3) Otsu segmentation
In the final section of this lab, we will be using Otsu segmentation to compute the binarization threshold. This is convenient when no prior knowledge of optimal thresholds is available or when there are fluctuations in signal-to-noise ratio from one image to another which makes the use of one static threshold (i.e: $\text{thr}=105$) unreliable for a large sequence of successive snapshots. It also offers a non-ambiguous, adaptive, and autonomous approach to computing optimal thresholds based on the histogram of each image. More info on the intuition and theory behind the algorithm is available in the original IEEE paper by Nobuyuki Otsu (DOI: 10.1109/TSMC.1979.4310076) or in the background section of [Chaib et al. 2023](https://doi.org/10.1007/s00348-023-03651-6).
\
\
The Otsu threshold is obtained by finding the intensity value $i$ which minimizes the intra-class variance or, equivalently, maximizes the inter-class variance. Although the former definition is more intuitive, the second will be used in this lab as it only based on first order statistical moments (i.e: class means) which results in order of magnitude decrease in computational overhead. Both definitions are equivalent and should yield the same value. In this lab, the Otsu threshold is defined as :
\
\
\begin{align}
thr  = \operatorname*{argmax}_{i}\ \omega_0 \omega_1 (\mu_0 - \mu_1)^2
\end{align}
\
where subscripts 0 and 1 refer to background ($i=<thr$) and foreground ($i>thr$) classes respectively. $\omega$ refers to the cumulative distribution function of the relevant class, or simply the proportion of pixels within it out of the total number of pixels in the image. $\mu$ refers to the mean of the relevant class.
A sample code with the algorithm's implementation is provided down below.

```python
# Determing the Otsu threshold
def otsu(img):
    img = 255*(img/np.max(img))
    img = img.astype(np.uint8)

    var = np.zeros(256)
    
    # Using the inter-class variance (first statistical moments only)
    for i in np.arange(1,255,1):
        # Thresholding the image using grayscale value i
        binarized_image = img>i
        # Computing conditional statistics
        cdf_foreground = np.sum(binarized_image)/np.size(img)
        cdf_background = 1-cdf_foreground
        mean_foreground = (np.mean(img[binarized_image == 1]))
        mean_background = (np.mean(img[binarized_image == 0]))
        # Computing the inter-class variance
        var[i] =  cdf_background*cdf_foreground*(mean_background-mean_foreground)**2
  
    # Determining the Otsu threshold (the last element is discarded due to empty foreground array)
    otsu_threshold = int(np.nanargmax(var[0:-2]))

    return otsu_threshold
```

```python
# Computing the Otsu threshold
thresh_otsu = otsu(img)
# Plotting the Otsu contour
y_manual,x_manual = np.where(contour_manual == 1)
img2 = (img>thresh_otsu).astype(int)
contour_otsu = img2 - cv2.erode(img2.astype(np.uint8),kernel = H)
y_otsu,x_otsu = np.where(contour_otsu == 1)
plt.figure(figsize=(5, 5))
plt.imshow(img, cmap='RdBu_r', vmin=0, vmax=255)
plt.colorbar(fraction=0.035, pad=0.04,ticks=[0, 255])
plt.xlabel("x [px]")
plt.ylabel("y [px]")
plt.title(img_id)
plt.plot(x_manual,y_manual,'.k',markersize = 1, label = "thr = 105 (manual)")
plt.plot(x_otsu,y_otsu,'.r',markersize = 1, label = f"thr = {thresh_otsu} (Otsu)")
plt.legend()
plt.rcParams.update({'font.size': 12})
plt.show()
```

Notice how our earlier prediction based on observation ($\text{thr}=105$) produces a flame front that is relatively close to the one obtained using the optimal Otsu threshold ($\text{thr}=96$) with a mean shift of the order of 1-2 pixels. When dealing with large datasets (i.e. 19,000+ images in Chaib et al. 2023), the Otsu threshold allows us to estimate the optimal threshold automatically for each image without having to inspect the histogram of each instantaneous snapshot (which would be very time consuming and impractical).
