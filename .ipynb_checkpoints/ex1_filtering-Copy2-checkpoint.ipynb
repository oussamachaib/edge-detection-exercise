# Lab 1: Image enhancement
## Introduction
In this first lab, we will introduce some of the conventional methods using for image enhancement. The goal at this stage is to improve the quality of the image so that the relevant features can be extracted easily. We will start with 1D signals for simplicity and move to 2D images later. We will mostly focus on the practical aspect while the theory behind each algorithm is presented in greater depth in some of the books and papers attached.\
Interesting tutorials can be found in the OpenCV website : https://docs.opencv.org/4.x/d2/d96/tutorial_py_table_of_contents_imgproc.html

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import cv2
import ipywidgets
```

```python
img = cv2.imread("data/sweeney1.png",0)
img = img.astype(np.uint8)
```

```python
plt.figure(figsize=(5,4))
plt.imshow(img, cmap='RdBu_r', vmin=0, vmax=255)
plt.colorbar(fraction=0.035, pad=0.04,ticks=[0, 255])
plt.xlabel("x [px]")
plt.ylabel("y [px]")
# LaTeX rendering
plt.rcParams.update({'font.size': 20})

plt.show()
```

```python
def plot_gaussian(w = 1):
    img_filt = cv2.GaussianBlur(img,(w,w),0)
    img_filt = img_filt/max(img_filt.flatten())
    plt.figure(figsize=(5,4))
    plt.imshow(img_filt, cmap='RdBu_r', vmin=0, vmax=1)
    plt.colorbar(fraction=0.035, pad=0.04,ticks=[0, 255])
    plt.xlabel("$x\,[px]$")
    plt.ylabel("$y\,[px]$")
    
    plt.show()
    
def plot_bilateral(w = 1):
    img_filt = cv2.bilateralFilter(img, d = w , sigmaColor=0, sigmaSpace=0)
    img_filt = img_filt/max(img_filt.flatten())
    plt.figure(figsize=(5,4))
    plt.imshow(img_filt, cmap='RdBu_r', vmin=0, vmax=1)
    plt.colorbar(fraction=0.035, pad=0.04,ticks=[0, 255])
    plt.xlabel("$x\,[px]$")
    plt.ylabel("$y\,[px]$")
    plt.itle("Gaussian filtering")
    plt.show()
    
def plot_grad1(w = 1):
    img_filt = cv2.GaussianBlur(img,(w,w),0)
    # Compute the gradient in the x direction
    gradient_x = cv2.Sobel(img_filt, cv2.CV_64F, 1, 0, ksize=w)
    # Compute the gradient in the y direction
    gradient_y = cv2.Sobel(img_filt, cv2.CV_64F, 0, 1, ksize=w)
    G = np.sqrt(gradient_x**2 + gradient_y**2)
    G = G/max(G.flatten())
    plt.figure(figsize=(5,4))
    plt.imshow(G, cmap='RdBu_r', vmin=np.min(G.flatten()), vmax=np.max(G.flatten()))
    plt.colorbar(fraction=0.035, pad=0.04,ticks=[0, 255])
    plt.xlabel("$x\,[px]$")
    plt.ylabel("$y\,[px]$")
    plt.show()
    
def plot_grad2(w = 1, sigma = 1):
    img_filt = cv2.bilateralFilter(img, d = w , sigmaColor=75, sigmaSpace=75)
    # Compute the gradient in the x direction
    gradient_x = cv2.Sobel(img_filt, cv2.CV_64F, 1, 0, ksize=w)
    # Compute the gradient in the y direction
    gradient_y = cv2.Sobel(img_filt, cv2.CV_64F, 0, 1, ksize=w)
    G = np.sqrt(gradient_x**2 + gradient_y**2)
    G = G/max(G.flatten())
    plt.figure(figsize=(5,4))
    plt.imshow(G, cmap='RdBu_r', vmin=np.min(G.flatten()), vmax=np.max(G.flatten()))
    plt.colorbar(fraction=0.035, pad=0.04,ticks=[0, 255])
    plt.xlabel("$x\,[px]$")
    plt.ylabel("$y\,[px]$")
    plt.show()
```

```python
ipywidgets.interact(plot_gaussian, w = (1,31,2))
ipywidgets.interact(plot_bilateral, w = (1,31,2)) # start, finish, step
```
